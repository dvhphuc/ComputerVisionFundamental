# **Process**

```python
def GraphSeg(path_filename, x, y, w, h):
  img = cv2.imread(path_filename)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img)
  plt.show()
  print(img.shape)

  mask = np.zeros(img.shape[:2], np.uint8)
  backgroundModel = np.zeros((1, 65), np.float64)
  foregroundModel = np.zeros((1, 65), np.float64)
  rectangle = (x, y, w, h)
  cv2.grabCut(img, mask, rectangle,
              backgroundModel, foregroundModel,
              5, cv2.GC_INIT_WITH_RECT)

  mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
  res = img * mask2[:, :, np.newaxis]
  plt.imshow(res)
  plt.show()
  res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
```

```python
GraphSeg('/content/drive/MyDrive/CS231.M21.KHTN/Baitap/Baitap6/images/Lionel-Messi.jpg',200,0,280,600)
```

```python
image = os.path.join(dir_image,'dogcat.jpg')
GraphSeg(image,10,10,140,170)
```
