# **Caculate Euclidean and Cosine**

```python
def TinhHist(pathfilename):
  img = cv.imread(pathfilename,0)
  hist = cv.calcHist([img],[0],None,
  [256],[0,256])
  size = img.shape[0]*img.shape[1]
  hist = hist / size
  return hist
```

```python
hist_image0814 = TinhHist('/content/gdrive/MyDrive/CS231.M21.KHTN/Baitap/Baitap1/images/image_0814.jpg')
hist_image0816 = TinhHist('/content/gdrive/MyDrive/CS231.M21.KHTN/Baitap/Baitap1/images/image_0816.jpg')
hist_image1248 = TinhHist('/content/gdrive/MyDrive/CS231.M21.KHTN/Baitap/Baitap1/images/image_1248.jpg')
```

```python
from scipy.spatial import distance
```

# Ảnh 0814 và 0816

```python
print('Euclidean : {}'.format(distance.euclidean(hist_image0814,hist_image0816)))
print('Cosine : {}'.format(1 - distance.cosine(hist_image0814,hist_image0816)))
```
